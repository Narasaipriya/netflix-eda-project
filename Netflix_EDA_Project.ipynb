```python
import pandas as pd
import matplotlib.pyplot as plt

netflix_df = pd.read_csv("netflix_data.csv")

# Display basic info
print("Dataset Shape:", netflix_df.shape)
print(netflix_df.head())
```

```python
# Keep only movies released between 1990 and 1999
movies_90s = netflix_df[
    (netflix_df['type'] == 'Movie') &
    (netflix_df['release_year'] >= 1990) &
    (netflix_df['release_year'] <= 1999)
]

print("\nTotal movies in 1990s:", movies_90s.shape[0])
```

```python
# Find the most common duration
duration = movies_90s['duration'].mode()[0]
print("\nMost frequent duration:", duration)
```

```python
# Filter action movies with duration less than 90 minutes
short_action_movies = movies_90s[
    (movies_90s['genre'] == 'Action') &
    (movies_90s['duration'] < 90)
]
# Count number of such movies
short_movie_count = short_action_movies.shape[0]
print("Short action movies count:", short_movie_count)

```

```python
# Movies Released Per Year 
movies_per_year = movies_90s['release_year'].value_counts().sort_index()

plt.figure()
movies_per_year.plot(kind='bar')
plt.title("Movies Released per Year (1990s)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
#Top Genres in 1990s
top_genres = movies_90s['genre'].value_counts().head(5)
print("\nTop Genres:\n", top_genres)

plt.figure()
top_genres.plot(kind='bar')
plt.title("Top Genres in 1990s")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
#Duration Distribution of Movies
plt.figure()
movies_90s['duration'].hist()
plt.title("Movie Duration Distribution (1990s)")
plt.xlabel("Duration")
plt.ylabel("Count")
plt.show()
```

```python
#Average Movie Duration
avg_duration = movies_90s['duration'].mean()
print("\nAverage duration:", round(avg_duration, 2))
```

```python
 #Long vs Short Movies Comparison
long_movies = movies_90s[movies_90s['duration'] >= 90].shape[0]
short_movies = movies_90s[movies_90s['duration'] < 90].shape[0]

print("\nLong movies:", long_movies)
print("Short movies:", short_movies)
```

## Insights from Netflix 1990s Analysis

- The most common movie duration is around 90–100 minutes, indicating a standard runtime pattern.
- Short action movies (<90 minutes) are relatively fewer compared to longer action films.
- Movie releases show an increasing trend towards the late 1990s.
- Drama and Action are the most dominant genres in the dataset.
- A majority of movies have a duration greater than 90 minutes, showing preference for full-length films.
